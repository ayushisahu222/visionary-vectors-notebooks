```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/perceptron_dataset.csv')
df.head()
```

### One-Hot Encoding - Converting text to numerical data.

```python
df['Client Demo'] = df['Client Demo'].map({'Yes': 1, 'No': 0})
df['Goes to Office'] = df['Goes to Office'].map({'Yes': 1, 'No': 0})  # This is your target column

def encode_workload(text):
    if "5 hrs job but can be done in focused 2 hours" in text:
        return 1  # Low workload
    elif "Trouble load" in text:
        return 2  # Medium workload
    elif "1 hour job but will take 7 hours" in text:
        return 3  # High workload

df['Workload'] = df['Workload Description'].apply(encode_workload)

df = pd.get_dummies(df, columns=['Weekday', 'Mood', 'Weather'], drop_first=True)

df = df.drop(columns=['Workload Description'])

df.head()
```
